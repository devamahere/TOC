```python
class stack :
    def __init__(self,l):
        self.l=l
    def push(self,l,a):
        self.l.append(a)
    def pop(self,l):
        self.l.pop()
    def top(self,l):
        print("top of stack is ",end=" ")
        print(self.l[len(l)-1])
    def display(self,l):
        print("your stack is ",end=" ")
        print(l)
l=[]
s=stack(l)
while True:
    print("enter 1 to push")
    print("enter 2 to pop")
    print("enter 3 to print top")
    print("enter 4 to display stack")
    print("enter 5 to exit")
    k=int(input())
    if k==1 :
        print("enter nuber to be pushed")
        b=int(input())
        s.push(l,b)
    elif k==2 :
        s.pop(l)
    elif k==3 :
        s.top(l)
    elif k==4 :
        s.display(l)
    elif k==5 :
        break
    else :
        print("enter a valid choice")
```
